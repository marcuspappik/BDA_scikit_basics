```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm, tree
```

```python
def plot_shape(clf, data):
    #create grid
    x = np.arange(data.x.min()-1, data.x.max()+1, 0.1)
    y = np.arange(data.y.min()-1, data.y.max()+1, 0.1)
    xx, yy = np.meshgrid(x, y)
    grid = pd.DataFrame({'x': xx.ravel(), 'y': yy.ravel()})
    
    #prdiction based on grid
    z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    z = z.reshape(xx.shape)
    
    #plot and color contour
    plt.plot(data.loc[data.class_label == 1, 'x'], data.loc[data.class_label == 1, 'y'], 'bo')
    plt.plot(data.loc[data.class_label == 0, 'x'], data.loc[data.class_label == 0, 'y'], 'ro')
    plt.contourf(xx, yy, z, cmap=plt.cm.RdYlBu, alpha=0.5)
    plt.show()
```

```python
data1 = pd.read_csv('data/svm_1.csv')
data2 = pd.read_csv('data/svm_2.csv')
data3 = pd.read_csv('data/svm_3.csv')
```

```python
#dataset 1
linear = svm.SVC(kernel='linear').fit(data1[['x', 'y']], data1['class_label'])
poly = svm.SVC(kernel='poly').fit(data1[['x', 'y']], data1['class_label'])
rbf = svm.SVC(kernel='rbf', gamma=2).fit(data1[['x', 'y']], data1['class_label'])
plot_shape(linear, data1)
plot_shape(poly, data1)
plot_shape(rbf, data1)
```

```python
#dataset 2
linear = svm.SVC(kernel='linear').fit(data2[['x', 'y']], data2['class_label'])
poly = svm.SVC(kernel='poly', degree=2).fit(data2[['x', 'y']], data2['class_label'])
rbf = svm.SVC(kernel='rbf', gamma=0.5).fit(data2[['x', 'y']], data2['class_label'])
plot_shape(linear, data2)
plot_shape(poly, data2)
plot_shape(rbf, data2)
```

```python
#dataset 3
linear = svm.SVC(kernel='linear').fit(data3[['x', 'y']], data3['class_label'])
poly = svm.SVC(kernel='poly', degree=2).fit(data3[['x', 'y']], data3['class_label'])
rbf = svm.SVC(kernel='rbf', gamma=10).fit(data3[['x', 'y']], data3['class_label'])
plot_shape(linear, data3)
plot_shape(poly, data3)
plot_shape(rbf, data3)
```

```python
#decission tree
tree_clf1 = tree.DecisionTreeClassifier().fit(data1[['x', 'y']], data1['class_label'])
tree_clf2 = tree.DecisionTreeClassifier().fit(data2[['x', 'y']], data2['class_label'])
tree_clf3 = tree.DecisionTreeClassifier().fit(data3[['x', 'y']], data3['class_label'])
plot_shape(tree_clf1, data1)
plot_shape(tree_clf2, data2)
plot_shape(tree_clf3, data3)
```
