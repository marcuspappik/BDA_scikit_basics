```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import math
import random
```

```python
x1 = np.arange(0, 15, 0.25)
y1 = np.sin(x1)*np.exp(np.sqrt(x1))
plt.plot(x1, y1, 'r')
plt.plot(x1, -y1, 'go')
plt.show()
```

```python
x2 = np.arange(1, 5, 0.1)
y2 = np.exp(x2)
plt.subplot(2, 1, 1)
plt.plot(x2, y2, 'ro')
plt.title('linear y axis')
plt.ylabel('y')
plt.xlabel('x')
plt.show()


plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'b^')
plt.title('log y axis')
plt.yscale('log')
plt.xlabel('same x')
plt.ylabel('another y')
plt.grid(True)
plt.show()
```
