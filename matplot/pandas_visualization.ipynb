```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#create data
x = np.array(range(50))
y1 = (2*np.array(x) - 25) - np.random.randint(-20, 20, 50)
y2 = (2*np.array(x) - 25)
df_plot = pd.DataFrame({'X': x, 'Y1': y1, 'Y2': y2})

#use dataframe own plot method
df_plot.plot()
plt.show()
```

```python
df_plot.loc[:, 'Xsq'] = x**2

#plot to columns against each other
df_plot.plot(x = 'Xsq', y = 'Y2')
plt.show()
```

```python
df_plot.loc[:,'Ydif'] = df_plot.loc[:,'Y1'] - df_plot.loc[:,'Y2']

#barplot of a certain column
df_plot.loc[:,'Ydif'].plot.bar()
plt.show()
```

```python
col1 = np.random.normal(3.5, 2, 1000)
col2 = np.random.normal(40, 20, 1000)
df = pd.DataFrame({'col1': col1, 'col2': col2})
```

```python
#scatter plot of two dimensions of dataframe
df.plot.scatter(x = 'col1', y = 'col2')
plt.show()
```
